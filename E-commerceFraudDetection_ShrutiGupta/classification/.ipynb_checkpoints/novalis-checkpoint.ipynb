```python
# Question 1

import pandas as pd
data = {'name':["Andy","Jose","Dieu","Chloe","Jose"], 'continent':["America","Europe","Asia","America","Europe"], 'college':["LibArts","Business","Engineering","Engineering","Business"]} 
  
# Creates pandas DataFrame. 
df = pd.DataFrame(data) 
  
# print the data 
df 
```

```python
# A dedup

df=df.drop_duplicates()
df
```

```python
# question B counts

COUNTS=df.groupby(['continent','college']).count()
print(COUNTS)
COUNTS=COUNTS.reset_index()
COUNTS
# grouped.shape
```

```python
#Q 1 C answer

PIVOTS=pd.pivot_table(COUNTS, values = 'name', index=['college'], columns = 'continent')
PIVOTS
```

```python
#Q1 f

zero_fill=PIVOTS.fillna(0)
zero_fill
```

```python
#Q1 f

zero_fill.sort_values('college', ascending=False)
```
